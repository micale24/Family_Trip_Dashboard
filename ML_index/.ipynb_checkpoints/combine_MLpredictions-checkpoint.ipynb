```python
from sklearn import tree
import pandas as pd
import numpy as np
import os
```

```python
safariML_df = pd.read_csv(os.path.join("output/safariML_predictions.csv"))
safariML_df.head()
```

```python
everestML_df = pd.read_csv(os.path.join("output/everestML_predictions.csv"))
everestML_df.head()
```

```python
dinoML_df = pd.read_csv(os.path.join("output/dinoML_predictions.csv"))
dinoML_df.head()
```

```python
date_df = pd.read_csv(os.path.join("data/date.csv"))
date_df.head()
```

```python
# Merge two dataframes using an inner join
combined_MLpredictions_df = pd.merge(safariML_df, everestML_df, on="dayofyear")
combined_MLpredictions_df
```

```python
# Merge two dataframes using an inner join
combined_MLpredictions_df = pd.merge(combined_MLpredictions_df, dinoML_df, on="dayofyear")
combined_MLpredictions_df
```

```python
# Merge two dataframes using an inner join
combined_MLpredictions_df = pd.merge(combined_MLpredictions_df, date_df, on="dayofyear")
combined_MLpredictions_df
```

```python
#Create clean metadata Dataframe with useful columns and rename "DATE" column
MLpredictions_df = combined_MLpredictions_df[["date","predicted_safari","predicted_everest","predicted_dino"]]
MLpredictions_df.head()
```

```python
MLpredictions_df = MLpredictions_df.astype({"predicted_safari": "int64", "predicted_everest": "int64", "predicted_dino": "int64"})
MLpredictions_df 
```

```python
# Export file as a CSV, without the Pandas index, but with the header
MLpredictions_df.to_csv("output/disney_MLpredictions.csv", index=False, header=True)
```
